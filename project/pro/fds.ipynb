```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
sns.set()
```

```python
data=pd.read_excel('fds1_proj.xlsx')
```

```python
data.info()
```

```python
data['weather'].unique()
```

```python
data['weather1']=np.nan
data['weather2']=np.nan
for i in range(0,len(data['weather'])):
    if data['weather'][i]=='Sunny':
        data['weather1'][i]=0
        data['weather2'][i]=1
    if data['weather'][i]=='Rainy':
        data['weather1'][i]=1
        data['weather2'][i]=0
    if data['weather'][i]=='Snowy':
        data['weather1'][i]=1
        data['weather2'][i]=1
```

```python
data.head()
```

```python
data['weather1']=data['weather1'].astype(int)
```

```python
data['weather2']=data['weather2'].astype(int)
```

```python
data.head()
```

```python
data['gender'].unique()
```

```python
data['gender1']=np.nan
for i in range(0,len(data['gender'])):
    if data['gender'][i]=='Male':
        data['gender1'][i]=0
    if data['gender'][i]=='Female':
        data['gender1'][i]=1
    
```

```python
data['gender1']=data['gender1'].astype(int)
```

```python
data.head()
```

```python
data['passanger'].unique()
```

```python
data['passanger_Alone']=np.nan
data['passanger_Friends']=np.nan
data['passanger_kids']=np.nan
data['passanger_Partners']=np.nan
for i in range(0,len(data['passanger'])):
    if data['passanger'][i]=='Alone':
        data['passanger_Alone'][i]=1
        data['passanger_Friends'][i]=0
        data['passanger_kids'][i]=0
        data['passanger_Partners'][i]=0
    if data['passanger'][i]=='Friend(s)':
        data['passanger_Alone'][i]=0
        data['passanger_Friends'][i]=1
        data['passanger_kids'][i]=0
        data['passanger_Partners'][i]=0
    if data['passanger'][i]=='Kid(s)':
        data['passanger_Alone'][i]=0
        data['passanger_Friends'][i]=0
        data['passanger_kids'][i]=1
        data['passanger_Partners'][i]=0
    if data['passanger'][i]=='Partner':
        data['passanger_Alone'][i]=0
        data['passanger_Friends'][i]=0
        data['passanger_kids'][i]=0
        data['passanger_Partners'][i]=1
```

```python
data.info()
```

```python
data['passanger_Alone']=data['passanger_Alone'].astype(int)
data['passanger_Friends']=data['passanger_Friends'].astype(int)
data['passanger_kids']=data['passanger_kids'].astype(int)
data['passanger_Partners']=data['passanger_Partners'].astype(int)
```
